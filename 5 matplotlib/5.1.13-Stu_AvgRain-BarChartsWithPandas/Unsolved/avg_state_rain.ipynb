```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
df = pd.read_csv('../Resources/avg_rain_state.csv')
df.head()
```

```python
# Set x axis and tick locations
x = df['State']
plt.figure(figsize=(20,4))
plt.xticks(ticks=np.arange(len(x)), labels=x, rotation='vertical')
y=df['Inches']
plt.xlim(0, len(x))
plt.ylim(0, max(df['Inches'])+10)
plt.xlabel('State')
plt.ylabel('Average Amount of Rainfall in Inches')
plt.title('Average Rain per State')
plt.bar(x,y, width=0.7, color='red', alpha=0.5, align='center');
plt.savefig('chart.png')
```
