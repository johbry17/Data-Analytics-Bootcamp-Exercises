```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
unemployed_data_one = pd.read_csv("../Resources/unemployment_2010-2015.csv")
unemployed_data_two = pd.read_csv("../Resources/unemployment_2016-2020.csv")

# Merge our two data frames together
# merged = pd.merge(unemployed_data_one, unemployed_data_two, on='Country Name')
merged = pd.merge(unemployed_data_one, unemployed_data_two, on=['Country Name', 'Country Code'])
merged
```

```python
# Delete the duplicate 'Country Code' column and rename the first one back to 'Country Code'
merged = merged.set_index('Country Code')
```

```python
# Set the 'Country Code' to be our index for easy referencing of rows
```

```python
# Collect the mean unemployment rates for the world
world = merged[[str(year) for year in range(2010, 2021)]].mean()
# Collect the years where data was collected
years = world.keys()
```

```python
# Plot the world average as a line chart
world_avg, = plt.plot(years, world, label='World Average', color='blue')
# Plot the unemployment values for a single country
country_avg, = plt.plot(years, merged.loc['USA', ["2010","2011","2012","2013","2014","2015",
                                    "2016","2017","2018","2019","2020"]], label='USA Average', color='green')

# Create a legend for our chart
plt.legend(handles=[world_avg, country_avg], loc='best')
# Show the chart
plt.show()
```

```python
# Plot the same data using the Pandas plot function
world.plot(label="World Average")
merged.loc['USA', "2010":"2020"].plot(label="United States")
plt.legend()
plt.show()
```
