```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
file = pd.read_csv('Resources/union_settlements_1995.csv')
file.head()
```

```python
# List columns
file.columns
```

```python
# Get total settlements by union
df = pd.DataFrame(file['UNION'].value_counts())
df
```

```python
# Configure plot, figsize, title, and axis labels
plot = df.plot(kind='bar', figsize=(10,6), color='r',  align='center')
plt.title("Major Collective Bargaining Settlements (1995)")
plt.ylabel('Settlements')
plt.xlabel('Union')
df = df.sort_index(axis=1)
# Configure x-tick rotation
plt.xticks(rotation=45, rotation_mode="anchor", ha='right', wrap=True)
# Show plot
plt.tight_layout()
```
