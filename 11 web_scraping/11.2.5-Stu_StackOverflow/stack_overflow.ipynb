```python
# Set the URL to visit to a variable
url = "https://stackoverflow.com/questions/tagged/python?sort=MostVotes&edited=true"

# Send the browser to the URL
browser.visit(url)

# Save the HTML from the browser
html = browser.html

# Create a BeautifulSoup object from the HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Find all divs that contain a question
qs = soup.find_all('div', class_="s-post-summary")
qs
```

```python
# Find all divs that contain a question
#soup.find_all('div', class_="s-post-summary js-post-summary")
```

```python
# Find the summary text of the first question
qs[0].find("a", class_="s-link").text
```

```python
# Find the vote count text of the first question
qs[0].find("span", class_="s-post-summary--stats-item-number").text
```

```python
# Find and strip the excerpt text of the first question
qs[0].find("div", class_="s-post-summary--content-excerpt").text.strip()
```

```python
# Create an empty list to store the questions
questions=[]
```

```python
# Create a for loop to extract the information
for _ in qs:
    # Extract the summary
    summary = _.find("a", class_="s-link").text
    
    # Extract the votes
    votes = _.find("span", class_="s-post-summary--stats-item-number").text
    
    # Extract the excerpt
    excerpt = _.find("div", class_="s-post-summary--content-excerpt").text.strip()
    
    # Create a dictionary to store the information
    thing = {'summary': summary, 'votes': votes, 'excerpt': excerpt}
    
    
    # Append the dictionary to the question list
    questions.append(thing)
```

```python
# Display the question list
questions
```

```python
# Close the browser
browser.quit()
```
