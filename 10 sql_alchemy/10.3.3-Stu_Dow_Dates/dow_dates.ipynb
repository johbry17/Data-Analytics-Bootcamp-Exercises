## Setup

```python
# Import dependencies.
import matplotlib
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, text, inspect, func
```

```python
engine = create_engine("sqlite:///Resources/dow.sqlite", echo=False)
conn = engine.connect()
```

```python
conn.execute(text('SELECT * FROM dow LIMIT 5')).fetchall()
```

```python
inspector = inspect(engine)
columns = inspector.get_columns('dow')
for c in columns:
    print(c['name'], c["type"])
```

## Reflect and query dates

```python
# Reflect Database into ORM class
Base = automap_base()
Base.prepare(autoload_with=engine)
Dow = Base.classes.dow
```

```python
s = Session(engine)
```

## Analysis

Analyze the Average prices (open, high, low, close) for all stocks in the Month of May

```python
# Query for the stock and average prices (open, high, low, close) 
# for all stock in the month of May
# Sort the result by stock name
# order_by is unnecessary, it's already alphabetical
_ = (
    s.query(
        Dow.stock,
        func.avg(Dow.open_price).label('open_avg'),
        func.avg(Dow.high_price).label('high_avg'),
        func.avg(Dow.low_price).label('low_avg'),
        func.avg(Dow.close_price).label('close_avg')
    )
    .filter(func.extract('month', Dow.date) == '05')
    .group_by(Dow.stock)
)

for row in _:
    print(row.stock, row.open_avg, row.high_avg, row.low_avg, row.close_avg, sep=', ')
```

```python
# Plot the Results in a Matplotlib bar chart
df = pd.DataFrame(_)

df.set_index('stock').plot(kind='bar', figsize=(10,6))
plt.xlabel('Stock')
plt.ylabel('Average Price')
plt.title('Average Prices for Stocks in May 2011')
plt.xticks(rotation=80)
plt.tight_layout()
plt.show()
```

### Bonus
Calculate the high-low peak-to-peak (PTP) values for `IBM` stock after `2011-05-31`. 
* Note: high-low PTP is calculated using `high_price` - `low_price`
* Use a DateTime.date object in the query filter
* Use list comprehension to create a list of dictionaries from the query results
* Create a dataframe from the list of dictionaries
* Use the `boxplot()` method on the dataframe to plot PTP distribution statistics

```python
# Design a query to calculate the PTP for stock `IBM` after May, 2011
_ = s.query(Dow.stock, Dow.high_price, Dow.low_price, Dow.date).filter(Dow.stock == 'IBM').filter(Dow.date > '2011-05-31')

for row in _:
    print(row.stock, row.high_price, row.low_price, row.date, sep=', ')

# List comprehension solution
_list = [{'stock':row.stock, 'high':row.high_price, 'low':row.low_price, 'date':row.date} for row in _]
_list
```

```python
# Load the list comprehension rows into a dataframe, set the index to the date, and plot the PTPs
df = pd.DataFrame(_list).set_index('date')

df['ptp'] = df.high - df.low
df.boxplot(column='ptp', flierprops={"markerfacecolor": "red", "markersize": 10})
plt.ylabel('Price to Price Difference')
plt.title('PTP Distribution for IBM Stocks for June 2011')
df
```
