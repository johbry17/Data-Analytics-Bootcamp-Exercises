# Australian Airports Map

```python
# Import the required libraries
import hvplot.pandas
import pandas as pd

# Turn off warning messages
import warnings
warnings.filterwarnings("ignore")
```

## 1. Load airports' data into a Pandas DataFrame

```python
# Create airport dataframe
df = pd.read_csv("../Resources/Airport_Output.csv")

# Display sample data
df
```

## 2. Create a simple map using GeoViews by adding a point per airport and setting a fixed size at your convenience

```python
# Configure the map
map1 = df.hvplot.points(
    "Lon",
    "Lat",
    geo = True,
    tiles = "OSM",
    frame_width = 800,
    frame_height = 600,
    color = "Airport Name",
    size=100,
    line_color="black",
    line_width=1.5,
)

# Display the map plot
map1
```
